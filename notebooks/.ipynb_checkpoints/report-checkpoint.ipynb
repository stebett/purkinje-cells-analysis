# General Parameters

```julia
data = load_data("data-v7.arrow")
plot_theme = Attributes(Axis = (xgridvisible = false, ygridvisible = false))
par = (
        linewidth = 2.,
        σ = 10,
        colormap = :vik,
        resolution = (2200, 1400),
        col_pos = ColorSchemes.RdBu_11[1],
        col_neg = ColorSchemes.RdBu_11[11],
        col_unmod = :black,
        col_mod = :gold,
        col_unmod_shade = RGBA(0,0,0,0.2),
        col_mod_shade = RGBA(1,0.,0.,0.25),
      )
```

## Figure 1

```julia
psth = PSTH((
             landmark = :cover,
             around = [-5000., 5000.],
             over = [-5000., -2000.],
             )...)

mpsth = MPSTH((
               around = 5000,
               num_bins = 20,
               b1 = 50,
               )...)

pnrate = PNRate((
             landmark = :cover,
             around = [-200., 200.],
             over = [-2000., -1000.],
             binsize = 50,
             )...)


psth_res = compute(psth, data)
mpsth_res = compute(mpsth, data)
pnrate_res = compute(pnrate, data)
```

```julia
psth_plot_params = (
                    kwargs = (colorrange = (0, 2), colormap = colormap),
                    sort_around = 800,
                    )

mpsth_plot_params = (
                     kwargs = (colorrange = (0, 2), colormap = colormap),
                     sort_binsize = 20,
                     )

```

```julia
fig = with_theme(plot_theme) do
    Figure(resolution = par.resolution)
end

psth_ax, psth_plot= visualise!(psth, fig, psth_res, psth_plot_params)
mpsth_ax = visualise!(mpsth, fig, mpsth_res, mpsth_plot_params)
pnrate_ax = visualise!(pnrate, fig, pnrate_res, psth_res, par)

fig[1, 1] = psth_ax
fig[1, 2] = mpsth_ax
fig[1, 3] = Colorbar(fig, psth_plot, colormap=colormap, width=10, label="Normalized change in firing rate")
fig[2, :] = pnrate_ax
hidespines!.([psth_ax, mpsth_ax, pnrate_ax], :t, :r)

label_a = fig[1, 1, TopRight()] = Label(fig, "A", textsize = 25, halign = :right)
label_b = fig[1, 2, TopRight()] = Label(fig, "B", textsize = 25, halign = :right)
label_c = fig[2, :, TopRight()] = Label(fig, "C", textsize = 25, halign = :right)

fig
```

```julia
save(plotsdir("report/figure1.png"), fig)
save("../docs/report/figures/analyses/figure1.png", fig)
```

## Figure 2

```julia
timecoursegroup = TimeCourseGroup((
                                   landmark = :cover,
                                   group = :n,
                                   around = [-150., 150.],
                                   )...)

correlations = Correlation((
                           around = [-150., 150.],
                           )...)



timecoursegroup_res = compute(timecoursegroup, data)
correlations_res = compute(correlations, data);
```

```julia
fig = with_theme(plot_theme) do
    Figure(resolution = resolution)
end

tc_ax1, tc_ax2, tc_ax3 = visualise!(timecoursegroup, fig, timecoursegroup_res, par)
corr_ax = visualise!(correlations, fig, correlations_res, par)

fig[1, 1] = tc_ax1
fig[2, 1] = tc_ax2
fig[3, 1] = tc_ax3
fig[:, 2] = corr_ax

hidespines!.([tc_ax1, tc_ax2, tc_ax3, corr_ax], :t, :r)

tc_ax1.title = "Firing time course around lift\nfor pairs of neighboring cells"
tc_ax2.ylabel = "Instantaneous firing rate"
tc_ax3.xlabel = "Time (ms)"

label_a = fig[1, 1, TopRight()] = Label(fig, "A", textsize = 25, halign = :right)
label_b = fig[1, 2, TopRight()] = Label(fig, "B", textsize = 25, halign = :right)

fig
```

```julia
save(plotsdir("report/figure2.png"), fig)
save("../docs/report/figures/analyses/figure2.png", fig)
```

```julia
cell1 = (
         index = 437,
         landmark = :cover,
         around = [-400., 400.],
         binsize = 0.5,
         )

cell2 = (
         index = 438,
         landmark = :cover,
         around = [-400., 400.],
         binsize = 0.5,
         )

neighcrosscor = GroupCrossCorr((
                    pad = 500,
                    num_bins = 2,
                    b1 = 25,
                    binsize = .5,
                    thresh_quant = 1.8,
                    group = couple(data, :n),
                    )...)

distcrosscor = GroupCrossCorr((
                    pad = 500,
                    num_bins = 2,
                    b1 = 25,
                    binsize = .5,
                    thresh_quant = 1.8,
                    group = couple(data, :d),
                    )...)


foldedcrosscor = FoldedCrossCorr((
                     σ=1,
                     binsize=0.5,
                     )...)


foldedcrosscorneigh = FoldedCrossCorrNeigh((
                                            σ=1
                                            )...)


rasters = [Raster(cell...) for cell in [cell1, cell2]]
timecourses = [TimeCourse(cell..., 10) for cell in [cell1, cell2]]



raster1_res = compute(rasters[1], data)
raster2_res = compute(rasters[2], data)
timecourse1_res = compute(timecourses[1], data)
timecourse2_res = compute(timecourses[2], data)
modcrosscor_res = compute(ModCrossCorr(), data)
neighbors_res = compute(neighcrosscor, data)
distant_res = compute(distcrosscor, data)
folded_res = compute(foldedcrosscor, data, neighbors_res)
foldedcrosscor_res = compute(foldedcrosscorneigh, data);
```

```julia
fig = with_theme(plot_theme) do
    Figure(resolution = resolution)
end

raster1_ax = visualise!(rasters[1], fig, raster1_res, par)
raster2_ax = visualise!(rasters[2], fig, raster2_res, par)
timecourse1_ax = visualise!(timecourses[1], fig, timecourse1_res, par)
timecourse2_ax = visualise!(timecourses[2], fig, timecourse2_res, par)
modcrosscor_ax = visualise!(ModCrossCorr(), fig, modcrosscor_res, par)
neigh_ax, dist_ax = visualise(neighcrosscor, fig, neighbors_res, distant_res, par)
folded_ax = visualise(foldedcrosscor, fig, folded_res, par)
folded_neigh_ax1, folded_neigh_ax2 = visualise(foldedcrosscorneigh, fig, foldedcrosscor_res, par)
fig[1, 1][1, 1] = raster1_ax
fig[1, 1][2, 1] = timecourse1_ax
fig[2, 1][1, 1] = raster2_ax
fig[2, 1][2, 1] = timecourse2_ax
fig[1, 2] = modcrosscor_ax
fig[2, 2] = folded_ax
fig[1, 3] = neigh_ax
fig[1, 4] = dist_ax
fig[2, 3] = folded_neigh_ax1
fig[2, 4] = folded_neigh_ax2

hidespines!.([raster1_ax, raster2_ax], :r, :t, :b)
hidespines!.([timecourse1_ax, timecourse2_ax, modcrosscor_ax, folded_ax, neigh_ax, dist_ax, folded_neigh_ax1, folded_neigh_ax2], :r, :t)
linkaxes!(neigh_ax, dist_ax)

fig[1, 1, TopRight()] = Label(fig, "A", textsize = 25, halign = :right)
fig[1, 2, TopRight()] = Label(fig, "B", textsize = 25, halign = :right)
fig[1, 3, TopRight()] = Label(fig, "C", textsize = 25, halign = :right)
fig[1, 4, TopRight()] = Label(fig, "D", textsize = 25, halign = :right)
fig[2, 1, TopRight()] = Label(fig, "E", textsize = 25, halign = :right)
fig[2, 2, TopRight()] = Label(fig, "F", textsize = 25, halign = :right)
fig[2, 3, TopRight()] = Label(fig, "G", textsize = 25, halign = :right)
fig[2, 4, TopRight()] = Label(fig, "H", textsize = 25, halign = :right)


fig
```

```julia
save(plotsdir("report/figure3.png"), fig)
save("../docs/report/figures/analyses/figure3.png", fig)
```
